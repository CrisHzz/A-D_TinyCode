Dada una matriz  arr de tamaño n  que contiene números enteros no negativos , la tarea es dividirla en dos conjuntos S1 y S2 de manera que la diferencia absoluta entre sus sumas sea mínima y encontrar la diferencia mínima.

Ejemplo

$array=[1,6,11,5]$

$Resultado:1$

```python
import numpy as np

def Absolute_Diference_Dynamic(array:list)->int:


  array_numpy:list=np.array(array)

  array_size:int=np.size(array_numpy)

  sum_elements:int=np.sum(array_numpy)

  arrayS=np.zeros((array_size+1,sum_elements+1),dtype=bool)

  arrayS[0][0]=True


  for i in range(1, array_size+1):
        arrayS[i] = arrayS[i-1]
        for j in range(sum_elements, array[i-1]-1, -1):
            arrayS[i][j] = arrayS[i][j] or arrayS[i-1][j-array[i-1]]

  for j in range((sum_elements+1)//2, -1, -1):
        if arrayS[array_size][j]:
            return abs(sum_elements - 2*j)


prueba:list=[1,6,11,5]

prueba2:list=[1,4]

print(Absolute_Diference_Dynamic(prueba))

print(Absolute_Diference_Dynamic(prueba2))
```
