# Software de Permutaciones,combinaciones,variaciones y Conteo

Primero antes que nada hay que entender un concepto matematico muy importante para entender este tipo de operaciones se trata de el **Factorial**

# Factorial

Se define como una operacion matematica indicada por el simbolo **!** que es el producto de toda la secuencia de numeros para atras siendo con un numero **n-1** hasta llegar al 1 ,.

Su ecuacion de formula como

$$ n!= (n-1)*(n-2)*(n-3)…*(n-n) $$

Existen algunos casos base que son excepcionales y hay que tener el cuenta

-No existen factoriales negativos

-Deben der numeros enteros

-El factorial de 0! es 1 por definicion

# Principios de conteo

Ya despues de entender que es el factoria sigue otros 2 conceptos interesantes de entender

**Principio de la Suma**: Cuando un evento puede ocurrir de $n_1$​ maneras y un segundo evento puede ocurrir de $n_2$​ maneras, y existe la condicion de que estos 2 eventos no pueden ocurrir al mismo tiempo, entonces la solucion de uno de los dos eventos es de $n_1​+n_2$​ maneras.

Tambien puede ser aplicado con el producto

**Principio del Producto**: Cuando un evento puede ocurrir de $n_1$​ maneras y un segundo evento puede ocurrir de $n_2$​ maneras, y los dos eventos pueden ocurrir al mismo tiempo, entonces los dos eventos pueden ocurrir es el resultado de la operacion $n_1​*n_2$​ maneras.


# Permutacion

Este concepto utiliza el factorial anteriormente presentado
y se enfonca en aquellas operaciones en donde se debe de cambiar el orden de algo , por lo general los ejercicios son de esee tipo, existen 2 variables muy vistas aqui que son

-n: Es el número total de elementos en el conjunto.

-r: Es el número de elementos que seleccionas del conjunto.

Tambien se implementan en aquellas operaciones que hay varios elementos repetidos mas letras del abecedario como S,T,U,V etc

**Permutacion ordinaria**:Son aquellas en donde solo hay 2 condiciones n el numero total de elementos del conjunto y r los elementos que se tomaran, no hay mas variables y su formula matematica se denota por

$$P(n,r)=\frac{n!}{(n-r)!}$$

Donde su numerador es el factorial de $n$ y su denominador es la el factorial de la resta de $n-$

**Permutacion Variada** = Su formula es muy parecida a la ordinaria solo que su resta desaparece y el numero de posibildiades que hay de un objeto a analizar son factoriales multiplicados entre si, por lo tanto su formula es

$$P=\frac{n!}{q!*r!*s!}$$

Si hay mas variables repetidas a analizar entonces se podran abajo y se multiplicaran entre si

# Combinaciones

Se define como la operacion en un conjunto de objetos que *no poseen un orden en especifico* donde $n$ es el número total de elementos en el conjunto y $r$ es el número de elementos que se seleccionan para formar grupos.

Su formula es tambien parecida a las anteriores y se define como

$$C(n,r)=\frac{n!}{(n-r)!*r!}$$

#Variaciones

Por ultimo tenemos las variaciones que  son una manera de seleccionar y ordenar elementos de un conjunto en especifico a analizar.

Hay dos tipos principales de variaciones:

-**Variaciones con repetición**:En este tipo de variaciones, los elementos pueden ser seleccionados más de una vez y el orden importa

$$n^{r}$$

-**Variaciones sin repetición**:En este tipo, los elementos se seleccionan sin reemplazo, lo que significa que cada elemento puede ser seleccionado solo una vez

$$\frac{n!}{(n-r)!}$$
